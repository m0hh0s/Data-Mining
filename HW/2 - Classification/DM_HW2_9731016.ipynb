## **FIRST STEP**

```python
import numpy as np
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
```

**First Thing we Need to Do is to Create Our Dataset**


```python
X, y = make_circles(n_samples=1000, factor=0.2, noise=0.1, random_state=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, random_state=42)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show
```

**Part 1:** Model Without Any Activation Function



```python
model_1 = keras.Sequential()

model_1.add(keras.layers.Dense(1, input_shape=(2,)))
model_1.add(keras.layers.Dense(1))

model_1.summary()
```

```python
model_1.compile(optimizer=tf.keras.optimizers.SGD(), loss='mse', metrics=['accuracy'])
history = model_1.fit(X_train, y_train, epochs=10)
```

As you can see above the accuracy is around **50%**, which is not acceptable.

**Part 2:** Model With Linear Activation Function

```python
model_2 = keras.Sequential()

model_2.add(keras.layers.Dense(1, activation='linear' ,input_shape=(2,)))
model_2.add(keras.layers.Dense(1, activation='linear'))

model_2.summary()
```

```python
model_2.compile(optimizer=tf.keras.optimizers.SGD(), loss='mse', metrics=['accuracy'])
history = model_2.fit(X_train, y_train, epochs=10)
```

Again the accuracy is below **50%**.

**Part 3:** Model With Linear Activation Function & Better Loss Function

```python
model_3 = keras.Sequential()

model_3.add(keras.layers.Dense(1, activation='linear' ,input_shape=(2,)))
model_3.add(keras.layers.Dense(1, activation='linear'))

model_3.summary()
```

```python
model_3.compile(optimizer=tf.keras.optimizers.SGD(), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
history = model_3.fit(X_train, y_train, epochs=100)
```

Even With 100 Epochs, the accuracy is not acceptable. Regression doesn't work because it draws a **Straight Line** through our data!

**Part 4:** Model With Only 1 Layer

```python
model_4 = keras.Sequential()

model_4.add(keras.layers.Dense(1, input_shape=(2,)))

model_4.summary()
```

```python
model_4.compile(optimizer=tf.keras.optimizers.SGD(), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
history = model_4.fit(X_train, y_train, epochs=10)
```

This model is not acceptable at all!

**Part 5:** Model With A Better Learning Rate

```python
model_5 = keras.Sequential()

model_5.add(keras.layers.Dense(1, activation='linear' ,input_shape=(2,)))
model_5.add(keras.layers.Dense(1, activation='linear'))

model_5.summary()
```

```python
model_5.compile(optimizer=tf.keras.optimizers.SGD(lr=0.01), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
history = model_5.fit(X_train, y_train, epochs=10)
```

Learning rate of 0.01 seems to be a better fit for this model. The accuracy reached **60%**, but still not good enough.

**Note:** If we set the LR too high, then the model tries to learn quickly but doesn't learn anything!

**Part 6:** The Ultimate Model :)) 

```python
model_6 = keras.Sequential([
  keras.layers.Dense(4, activation='relu'),
  keras.layers.Dense(4, activation='relu'),
  keras.layers.Dense(1, activation='sigmoid')               
])

model_6.compile(optimizer=tf.keras.optimizers.Adam(lr=0.1), loss=tf.keras.losses.BinaryCrossentropy(), metrics=['accuracy'])
history = model_6.fit(X_train, y_train, epochs=20)
```

**WOW!!!**

**100%** accuracy is achieved!!!

The keys to this model are:

*    Activation functions which are **non-linear**. They fit perfectly with our data.
*   Using the **Adam optimizer** instead of SGD.
*   Setting a **Higher LR** combined with **Fewer Epochs**.

```python
loss, accuracy = model_6.evaluate(X_test, y_test)
```

It also runs perfectly on the test set.

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

**Loading Our Dataset:**


```python
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()
```

**Data Preprocessing:**


```python
train_images = train_images / 255.0
test_images = test_images / 255.0
```

**Creating & Compiling Our Model:**


```python
model = tf.keras.Sequential([
                             tf.keras.layers.Flatten(input_shape=(28, 28)),
                             tf.keras.layers.Dense(128, activation='relu'),
                             tf.keras.layers.Dense(10)
                             
])

model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics='accuracy')
```

**Training Our Model:**


```python
model.fit(train_images, train_labels, epochs=10)
```

Almost 91% accuracy on train set!!!

```python
loss, accuracy = model.evaluate(test_images, test_labels)
```

88% accuracy on test set

**Creating & Plotting The Confusion Matrix:**


```python
pred = utils.label_with_highest_prob(model.predict(test_images))

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

utils.plot_confusion_matrix(y_pred=pred,
                            y_true=test_labels,
                            classes=np.array(class_names))
```
